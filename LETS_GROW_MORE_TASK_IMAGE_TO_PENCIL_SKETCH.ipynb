# IMAGE TO PENCIL SKETCH USING PYTHON

# importing the libraries

```python
import cv2
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

# reading the image

```python
img = cv2.imread("C:\parrot.jpg")
img = cv2.resize(img,(480,270))
i_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(i_rgb)
```

# converting the original image to greyscale:

```python
from IPython.display import display
greyimg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(greyimg))
```

# inverting the new grayscale image:

```python
invtimg = cv2.bitwise_not(greyimg)
display(Image.fromarray(invtimg))
```

#  bluring the image by using the Gaussian Function in OpenCV:

```python
blurimg = cv2.GaussianBlur(invtimg,(5,5),0,0)
display(Image.fromarray(blurimg))
```

#  Creating the Pencil Sketch


```python
invtblurimg = 255 - blurimg
sketch = cv2.divide(greyimg,invtblurimg,scale=250.0)
display(Image.fromarray(sketch))
```

# Comparing the Original Image with Pencil Sketch

```python
display(Image.fromarray(i_rgb))
display(Image.fromarray(sketch))
```
